# Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Leitura do dataset

```python
path = 'C:/Users/jjoao/AnaliseDeDados/'
```

```python
df = pd.read_excel(path + 'meteorite_landings.xlsx')
df.head()
```

```python
df_new = df[df['reclong'] < 300]
```

```python
plt.hist(df_new['reclat'])
plt.show()
```

## Variando o número de bins

```python
plt.hist(df_new['reclat'], bins = 5)
plt.show()
```

```python
plt.hist(df_new['reclong'])
plt.show()
```

```python
plt.figure(figsize = (15, 5))
plt.subplot(1,2,1)
plt.hist(df_new['reclong'])
plt.xlabel('Longitude')
plt.ylabel('Frequência')
plt.subplot(1,2,2)
plt.hist(df_new['reclat'], color = 'green')
plt.xlabel('Latitude')
plt.ylabel('Frequência')
plt.show()
```

## Latitude e Longitude no mesmo gráfico

```python
plt.figure(figsize = (15,5))
plt.hist([df['reclat'], df['reclong']], bins = 50, label=['lat', 'long'])
plt.legend(loc = 'upper right')
plt.show()
```

## Variando o número de bins

```python
plt.figure(figsize = (15,5))
plt.hist([df['reclat'], df['reclong']], bins = 20, label=['lat', 'long'])
plt.legend(loc = 'upper right')
plt.show()
```
